```python
import pandas as pd
from sqlalchemy import create_engine

# Local MySQL se connect karne ke liye
engine = create_engine("mysql+pymysql://root:%40Enter.com@localhost/hr_analytics")

# Table se data read karo
df = pd.read_sql("SELECT * FROM employees", con=engine)

# Data check karo
df.head()
```

```python
print(df.isnull().sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(6,4))

sns.countplot(x="Attrition",data=df,palette='Set2')

plt.title("Total Attrition Count in Company")
plt.xlabel("Attrition")
plt.ylabel("Numbers of Employees")

plt.show()
```

```python
plt.figure(figsize=(7, 5))

# Overtime aur Attrition ka comparison plot
sns.countplot(x='OverTime',hue='Attrition',data=df,palette='Set1')

plt.title('Impact of Overtime on Attrition')
plt.xlabel('Does Employee do Overtime?')
plt.ylabel('Number of Employees')
plt.show()
```

```python
plt.figure(figsize=(6, 5))

# Attrition ke hissab se Monthly Income ka boxplot
sns.boxplot(x='Attrition', y='MonthlyIncome', data=df, palette='Set2')

plt.title('Monthly Income vs Attrition')
plt.xlabel('Attrition (Chhoda ya Nahi?)')
plt.ylabel('Monthly Income (Salary)')
plt.show()
```

```python
plt.figure(figsize=(12, 8))

# Sirf numeric columns filter kar rahe hain kyunki correlation numbers pe nikalta hai
numeric_df = df.select_dtypes(include=['int64', 'float64'])

# Heatmap plot karna
sns.heatmap(numeric_df.corr(), cmap='coolwarm', annot=False)

plt.title('Correlation Heatmap of Employee Features')
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Horizontal bar chart taaki Job Roles ke naam saaf dikhein
sns.countplot(y='JobRole', hue='Attrition', data=df, palette='Set2')

plt.title('Job Role wise Attrition Rate')
plt.xlabel('Number of Employees')
plt.ylabel('Job Role')
plt.show()
```
